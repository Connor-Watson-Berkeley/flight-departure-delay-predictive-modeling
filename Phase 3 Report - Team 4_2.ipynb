Flight delays cost airlines billions annually through crew scheduling disruptions, gate misallocations, and passenger compensation while frustrating travelers through missed connections and uncertainty. This multi-phase project aims to predict domestic flight departure delays in minutes two hours before scheduled departure, enabling proactive airline operations and improved passenger experience.

Using the U.S. Department of Transportation (DOT) Bureau of Transportation Statistics On-Time Performance dataset (covering over 70 million flights from 2015–2021), integrated with NOAA Integrated Surface Database weather observations and airport metadata, we identify key delay patterns and build a predictive modeling framework. Exploratory data analysis revealed that while 80% of flights operate within 15 minutes of schedule, delays are highly variable (mean 10 minutes, σ = 38), with elevated occurrences on Mondays, Sundays, and during peak summer months. These findings highlight the importance of temporal, network, and weather-related predictors for delay forecasting.

In Phase 1, we chose to frame this as a regression problem predicting continuous delay times, evaluated using RMSE which provides an interpretable measure of average prediction error in minutes that can be directly understood by stakeholders. Predictions were post-processed into delay buckets and assessed using the domain-specific metrics: On-Time Performance Prediction Accuracy and Severe Delay Detection Rate. These metrics aim to assess both practical and statistical impacts.

For Phase 2, we established a baseline linear regression model using a robust 6-stage feature engineering pipeline encompassing data preparation, median imputation, categorical encoding, feature assembly, standardized scaling, and model training. We selected linear regression as our baseline for its interpretability, computational efficiency on distributed systems, and ability to establish clear performance benchmarks for subsequent model comparisons. Using expanding window cross-validation across five temporal folds, our baseline model achieved a test set RMSE of 35.96 minutes

### The Fundimental Equation

$$\text{Delay} = \text{Available Time} - \text{Air Time} - \text{Turnover Time}$$

Where:
- **Available Time**: Time between previous flight's departure and current flight's scheduled departure
- **Air Time**: Actual flight duration (wheels-off to wheels-on)
- **Turnover Time**: Ground operations time (previous arrival to current departure)

When **Available Time < (Air Time + Turnover Time)**, the current flight departs late.

### Delay Propagation Mechanisms

1. **Late Arrivals Lead to Late Departures**: If a previous flight arrives late, the aircraft has less time for ground operations, increasing the likelihood that the current flight will depart late.

2. **Operational Time Variability Compounds**: Variability in air time (weather, air traffic) and turnover time (gate availability, crew scheduling, maintenance) accumulates through the rotation, creating cascading delays.

3. **Network Effects Amplify Impact**: Delays at hub airports propagate through connecting flights, affecting multiple downstream flights in the network.

### Application to Predictive Modeling

The primary motivation for our feature engineering strategy is to provide our models the information they need in order to estimate this equation. To capture cascading delay effects, we need information about previous flight operations. However, actual duration features contain data leakage (they use future information). Our multi-pronged approach includes:

1. **Safe Features**: Intelligent data leakage handling that uses actual values when available at prediction time, and falls back to scheduled times or estimates when not available.

2. **Graph Analysis Features**: PageRank-based network centrality metrics that capture airport congestivity and network effects, providing insights into delay propagation patterns.

3. **Meta-Models**: Train separate models to predict previous flight durations (air time, turnover time, rotation time) using only information available at prediction time, then use these predictions as features in our main departure delay model.

### Flight Lineage Features

Flight lineage features capture the cascading effects of delays through aircraft rotations by joining each flight to its previous flight in the same aircraft's daily sequence. Combined with temporal features that capture time-of-day, day-of-week, and seasonal patterns, these features provide critical information for predicting departure delays.

#### Key Terminology

- **Air Time**: Time from wheels-off to wheels-on (actual flight duration, excluding all ground operations)
- **Turnover Time**: Time from previous flight's actual arrival to current flight's actual departure (includes taxi-in, gate operations, and taxi-out)
- **Rotation Time**: Time from previous flight's actual departure to current flight's actual departure (complete cycle = Air Time + Turnover Time)

These terms map directly to our fundamental theory of delay propagation, where **Delay = Available Time - Air Time - Turnover Time**.

### Connection to Fundamental Theory

As established in our fundamental theory of delay propagation, delays cascade through the airline network via aircraft rotations. When an aircraft experiences delays on one flight, those delays propagate to subsequent flights in the same aircraft's daily rotation. 

**The Challenge**: To capture these cascading delay effects in our models, we need information about previous flight operations. However, **we cannot directly use actual duration features** because they may contain data leakage.

### The Solution: Meta-Models

Instead of using actual durations (which leak future information), we train **meta-models** to **predict** these durations based on features that are available at prediction time:

1. **Scheduled times**: Scheduled rotation time, scheduled turnover time, and scheduled flight time provide the baseline expectations for how long operations should take according to the carrier's schedule.

2. **Temporal patterns**: Month, day of week, and hour of day (departure time block) capture diurnal and seasonal patterns, including jet stream variations that affect flight duration throughout the day and year.

3. **Weather conditions**: Wind speed/direction, precipitation, visibility, temperature, pressure, and humidity at origin and destination airports (available at prediction time) directly impact flight duration through headwinds, turbulence, and approach restrictions.

4. **Airport and Trip characteristics**: Origin and destination airports, route distance, airport elevation, and network centrality (PageRank) capture route-specific and airport-specific operational patterns that influence duration.

We can predict previous flight durations using only information available at prediction time, then use these predictions as features in our main departure delay model. This gives us the predictive power of duration features without the data leakage, enabling us to capture the cascading delay effects described in our fundamental theory.

### Model Selection

We considered several modeling approaches to prove the meta-model strategy:

1. **Linear Regression**: Too simple (high bias) - would not capture the complex non-linear relationships between features and flight durations.

2. **Neural Network**: Would likely take too long to run/evaluate and require significant effort to tune (high variance risk). Given this was a proof-of-concept effort, the tuning overhead was prohibitive.

3. **Random Forest**: Offers a good balance of bias and variance, with strong predictive performance out-of-the-box requiring very minimal hyperparameter tuning. This was our initial choice.

4. **XGBoost**: Selected after benchmarking showed similar performance to Random Forest but with dramatically faster training times.

###Loss Function:
####  Mean Squared Error (MSE)
Our models predict continuous flight delay times in minutes using Mean Squared Error (MSE) as its loss function during training, which penalizes large prediction errors more heavily than small ones. It is useful as our loss function as it measures the average squared difference between predicted and actual delay times. MSE penalizes large errors more than small errors due to the squaring operation, making it sensitive to outliers and ensuring the model focuses on reducing substantial prediction errors.

Mean Squared Error (MSE), calculated as 

$$MSE = \frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y}i)^2$$

| Loss Function | Formula | Description |
|--------|---------|-------------|
| Mean Squared Error (MSE) | $$\mathrm{MSE} = \frac{1}{N} \sum_{i=1}^{N} (y_i - \hat{y}_i)^2$$ | Mean Squared Error - average of squared prediction errors, useful for optimization but less interpretable than RMSE |

###Key Regression Metrics
#### Root Mean Squared Error (RMSE)
We've selected RMSE (the square root of MSE) as one of our key regression metrics as it is expressed in the same units as the target variable (minutes), providing a more interpretable measure of average prediction error magnitude than MSE which can be directly understood by stakeholders. 

The Root Mean Squared Error (RMSE), computed as

$$RMSE = \sqrt{MSE} = \sqrt{\frac{1}{n}\sum{i=1}^{n}(y_i - \hat{y}i)^2}$$

| Metric | Formula | Description |
|--------|---------|-------------|
| Root Mean Squared Error (RMSE) | $$\mathrm{RMSE} = \sqrt{\frac{1}{N} \sum_{i=1}^{N} (y_i - \hat{y}_i)^2}$$ | Root Mean Squared Error - measures average prediction error magnitude in original units, penalizes large errors more heavily |

#### Mean Absolute Error (MAE)

Additonally, we assess our models on Mean Absolute Error (MAE), defined as

$$
MAE = \frac{1}{n}\sum{i=1}^{n}|y_i - \hat{y}i|
$$

calculates the average absolute difference between predicted and actual delays without squaring, making it less sensitive to outliers than MSE or RMSE and providing a robust measure of typical prediction error.

|--------|---------|-------------|

###Key Classification Metrics
Beyond standard regression metrics, we evaluate operational utility through domain-specific classification metrics to help business stakeholders understand our models prediction outcomes.

#### On-Time Prediction Accuracy (OTPA)
OTPA (On-Time Prediction Accuracy) measures binary classification accuracy for the critical 15-minute on-time threshold defined by the U.S. Department of Transportation.

**On-Time Prediction Accuracy** measures the accuracy of predicting flights within a narrow tolerance window (+/- 5 minutes), which is critical for operational planning because resource allocation decisions, such as gate assignments, crew scheduling, and passenger communication, require precise delay estimates rather than broad categorizations. 

OTPA converts delay predictions into a binary outcome—*on-time* (delay < 15 minutes) or *delayed* (≥ 15 minutes)—and measures accuracy or F1-score. This bridges regression and classification perspectives, aligning with how the aviation industry defines on-time performance. On-Time flights help improve the reputation of the airline and keep the airport traffic smoothly moving. 

We operationalize this metric by converting continuous delay predictions into binary outcomes consistent with U.S. Department of Transportation (DOT) definitions:

- On-Time: delay < 15 minutes
- Delayed: delay ≥ 15 minutes

**On-Time Definition:**

- Predicted On-Time = 1 if `ŷ < 15`, else 0  
- Actual On-Time = 1 if `y < 15`, else 0  

**Formula:**

`On-Time Prediction Accuracy (OTPA) = (TP + TN) / (TP + TN + FP + FN)`

Where:  
- **TP** = correctly predicted on-time flights  
- **TN** = correctly predicted delayed flights  
- **FP** = predicted on-time but actually delayed  
- **FN** = predicted delayed but actually on-time  

| Metric | Formula | Description |
|--------|---------|-------------|
| **On Time Prediction Accuracy (OTPA)** | $$\text{OTPA} = \frac{TP + TN}{TP + TN + FP + FN}$$ | Binary classification accuracy for on-time (<15 min) vs delayed (≥15 min) flights |

This metric translates regression outputs into actionable insights for **crew scheduling**, **resource allocation**, and **passenger notifications**, emphasizing precision in predicting near-schedule flights.

#### Severe Delay Detection Rate (SDDR)
Severe Delay Detection Rate measures the recall specifically for delays >=60 minutes, which is important for identifying flights requiring major operational adjustments such as aircraft substitutions, crew reassignments, or passenger compensation, as these severe delays have the most significant impact on operations and customer satisfaction. 

Buckets:
- OnTime: < 15 minutes
- Delayed: 15-60 minutes
- Severe: 60+ minutes

Formula:

Severe Delay Detection Rate (SDDR) = TP_severe / (TP_severe + FN_severe)

Where:

- **TP_severe** = number of flights correctly predicted to have severe delays (≥ 60 min)
- **FN_severe** = number of severely delayed flights the model failed to identify

| Metric | Formula | Description |
|--------|---------|-------------|

## Appendix
Regression Metrics
| Metric | Formula | Description |
|--------|---------|-------------|
| RMSE | $$\mathrm{RMSE} = \sqrt{\frac{1}{N} \sum_{i=1}^{N} (y_i - \hat{y}_i)^2}$$ | Root Mean Squared Error - measures average prediction error magnitude in original units, penalizes large errors more heavily |
| MAE | $$\mathrm{MAE} = \frac{1}{N} \sum_{i=1}^{N} \|y_i - \hat{y}_i\|$$ | Mean Absolute Error - measures average absolute prediction error, more robust to outliers than RMSE |
| R² | $$R^2 = 1 - \frac{\sum_{i=1}^{N} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{N} (y_i - \bar{y})^2}$$ | Coefficient of Determination - proportion of variance in the target variable explained by the model (0-1 scale) |
| MSE | $$\mathrm{MSE} = \frac{1}{N} \sum_{i=1}^{N} (y_i - \hat{y}_i)^2$$ | Mean Squared Error - average of squared prediction errors, useful for optimization but less interpretable than RMSE |

Classification Metrics
| Metric | Formula | Description |
|--------|---------|-------------|
| **OTPA Accuracy** | $$\text{OTPA} = \frac{TP + TN}{TP + TN + FP + FN}$$ | Binary classification accuracy for on-time (<15 min) vs delayed (≥15 min) flights |
| **OTPA Recall** | $$\text{Recall} = \frac{TP}{TP + FN}$$ | Proportion of actual on-time flights correctly identified |
| **SDDR (Recall)** | $$\text{SDDR} = \frac{TP_{severe}}{TP_{severe} + FN_{severe}}$$ | Proportion of severe delays (≥60 min) correctly identified - critical for operational planning |
| **Bucket Accuracy** | $$\text{Accuracy} = \frac{\text{Correct Predictions}}{\text{Total Predictions}}$$ | Overall accuracy across all 4 delay buckets (Early, OnTime, Delayed, Severe) |
